# Assign IDs to Alphabet and Punctuators

The IDs are not meaningful in and of themselves, but allow us to match letters/puncts with 
other associated metadata.

```python
from pathlib import Path
import json
```

```python
alphabet_json = Path('../alphabet/alphabet.json') 
punct_json = Path('../punctuation/punctuation.json')

alpha_data = json.loads(alphabet_json.read_text())
punct_data = json.loads(punct_json.read_text())
```

```python
new_alpha_data = {}
new_punct_data = {}

ID = 1
for dtype in (alpha_data, punct_data):
    for d in dtype:
        new_alpha_data[ID] = d
        ID += 1
```
